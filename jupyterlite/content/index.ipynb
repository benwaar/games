# Utala: KAOS 9

## A Competitive 2-Player Tactical Duel

Welcome to **Utala: KAOS 9** — a strategic card game playable with any standard 52-card deck!

This is an interactive research project exploring game AI through multiple agent types:
- **Random** baseline
- **Heuristic** strategic agents
- **Monte Carlo** search-based agents
- **DQN** deep reinforcement learning agents

Everything is **text-based, inspectable, and hackable**.

## Game Rules

**Utala: KAOS 9** is a tactical 3x3 grid combat game:

1. **Placement Phase**: Take turns placing your 9 rocketmen (cards 2-10) on the board
   - Cards 2, 3, 9, 10 are placed face-down for deception
   - Contest opponent squares for tactical advantage

2. **Dogfight Phase**: Resolve contested squares through turn-based combat
   - Center square resolves first, then the winner chooses the next square
   - Use your 4 weapons (Ace, King, Queen, Jack) strategically
   - Draw Kaos cards for resolution

3. **Win Condition**: Get 3 in a row, or control the most squares

## About This Project

This is a game AI research project spanning 5 completed phases:
- **Phase 1**: Engine, replay, baselines (Random, Heuristic, Monte Carlo)
- **Phase 2**: TD-Linear learning with hand-built gradients
- **Phase 3**: DQN deep learning with imitation distillation
- **Phase 4**: Rule evolution (v1.9 choosable dogfight order)
- **Phase 5**: Improved DQN, distilled production model, shipped to Flutter app

**Project ethos:** Understanding algorithms by building them, seeing how different approaches *think*, learning by stripping problems down to their **nuts and bolts**.
